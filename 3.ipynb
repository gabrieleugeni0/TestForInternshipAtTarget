```python
import pandas as pd
```

```python
df = pd.read_json('dados.json')
df
```

```python
df_clear = df.query('valor != 0')
df_clear
```

```python
menor_faturamento = df_clear.loc[df_clear.valor == df_clear.valor.min()]
print(f'O menor valor de faturamento foi de {menor_faturamento.valor.values[0]}, ocorrido no dia {menor_faturamento.dia.values[0]}.')
```

```python
maior_faturamento = df_clear.loc[df_clear.valor == df_clear.valor.max()]
print(f'O menor valor de faturamento foi de {maior_faturamento.valor.values[0]}, ocorrido no dia {maior_faturamento.dia.values[0]}.')
```

```python
media_faturamento = df_clear.valor.mean()
dias_maior_que_media = df_clear.loc[df_clear.valor > media_faturamento, 'valor'].count()
print(f'A media de faturamento mensal foi de {media_faturamento:.2f}, sendo que em {dias_maior_que_media} dias o faturamento diario foi maior que esse valor medio.')
```
